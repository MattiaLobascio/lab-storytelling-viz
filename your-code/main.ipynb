```python
# your code here
# add cells as needed

# Plot correlation between Sale Price and Size of the House (General Living Area,Garage Area,Lot Area))
# Last plot examines correlation between General Living Area and Lot Area
fig, ax = plt.subplots(2,2, figsize=(20, 10))
ax[0,0].scatter(df["SalePrice"],df["GrLivArea"])
ax[1,0].scatter(df["SalePrice"],df["GarageArea"])
ax[0,1].scatter(df["SalePrice"],df["LotArea"])
ax[1,1].scatter(df["GrLivArea"],df["LotArea"])
ax[0,0].set_xlabel("Sale Price")
ax[0,0].set_ylabel("General Living Area")
ax[1,0].set_xlabel("Sale Price")
ax[1,0].set_ylabel("Garage Area")
ax[0,1].set_xlabel("Sale Price")
ax[0,1].set_ylabel("Lot Area")
ax[1,1].set_xlabel("General Living Area")
ax[1,1].set_ylabel("Lot Area")
```

```python
#Group Kitchen and Bathrooms in Service Rooms column
df["ServiceRooms"]  = df["KitchenAbvGr"]+df["FullBath"]+df["HalfBath"]+df["BsmtFullBath"]+df["BsmtHalfBath"]

#Plot the dispersion of the price based on number of Rooms and Service Rooms
fig, ax = plt.subplots(1,2, figsize=(16, 8))

plt.sca(ax[0])
sns.boxplot(x="BedroomAbvGr", y="SalePrice", data=df)
plt.sca(ax[1])
sns.boxplot(x="ServiceRooms", y="SalePrice", data=df)

#In both the cases we have a relevant number of outliers.
```

```python
df["Age"] = (2019 - df["YearBuilt"])
df["Renovation"] = (2019 - df["YearRemodAdd"])

#Plot the distribution of the price based on Age of the house and Years since last renovation
fig, ax = plt.subplots(1,2, figsize=(16, 8))

plt.sca(ax[0])
sns.lineplot(x="Age", y="SalePrice", data=df)
plt.sca(ax[1])
sns.lineplot(x="Renovation", y="SalePrice", data=df)

#Apart for some isolated cases (probably prestigious old houses), the higher the age/time since
#last renovation, the lower the Sale Price
```

```python
#group by neighborhood
nb = df.groupby(by="Neighborhood").agg({"SalePrice":["count","mean","median"]})

#Plot relationship between Saleprice and neighborhood plus number of deals per neighborhood
fig,ax = plt.subplots(1,2, figsize=(18,8))

plt.sca(ax[0])
sns.boxplot(x="SalePrice", y="Neighborhood", data=df)

plt.sca(ax[1])
plt.barh(nb.index,nb["SalePrice"]["count"].sort_values(ascending=True))
plt.xlabel("Number of Deals")
plt.ylabel("Neighborhood")

```

## Bonus Challenge 2 - Exploring Data with Correlation Heatmap

Now you have explored data visualizations with certain fields based on common sense. In the dataset there are many other fields that you are not sure whether they are important factors for the sale price. What is the best way to explore those fields without investigating them individually?

Making scatter matrix is not an option here because there are too many fields which makes it extremely time consuming to create scatter matrix. One option you have is to create a heatmap. Heatmaps are much less expensive to create than scatter matrixes. You can use heatmaps to visualize the pairwise correlations between each two variables.

Here is a [reference](https://seaborn.pydata.org/examples/many_pairwise_correlations.html) you can use to learn how to creat the pairwise correlation heatmap. Your heatmap should look like below:

![Corr Heatmap](heatmap.png)

```python
# your code here
sns.set(style="white")

# Compute the correlation matrix
corr = df.corr()

# Generate a mask for the upper triangle
mask = np.zeros_like(corr, dtype=np.bool)
mask[np.triu_indices_from(mask)] = True

# Set up the matplotlib figure
f, ax = plt.subplots(figsize=(11, 9))

# Generate a custom diverging colormap
cmap = sns.diverging_palette(220, 10, as_cmap=True)

# Draw the heatmap with the mask and correct aspect ratio
sns.heatmap(corr, mask=mask, cmap=cmap, vmax=1, center=0, vmin=-1,
            square=False, linewidths=.5)
```

In your heatmap, you can easily identify the highly correlated (either positively or negatively) variables by looking for the grids with darker colors.
